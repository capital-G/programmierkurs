## Scope

Der *Scope* einer Funktion gibt an ob wir Zugriff auf eine Variable haben oder nicht.

In Python hat man in der Regel nur Zugriff auf die Variablen aus der kleineren Einrckung und die ber der Zeile stehen.

````python
foo = 42

if bar is True:
    print(foo)

bar = True
````

````python
foo = 'Hello'

def my_func():
    print(foo)

my_func()
````

````python
del foo

def my_func():
    foo = 42
    bar = 2
    return bar

print(foo)
````

## PEP8

Wenn mehrere Leute an einem Projekt arbeiten ist es sinnvoll sich an sogennante *Style Guidelines* zu halten, die die Formatierung wie Namen von Variablen, Funktionen oder Art der Einrckungen vereinheitlichen.

Bei Python nennt sich dieser Standard *pep8* und ist hier sepzifiziert:
https://www.python.org/dev/peps/pep-0008/

Art | Konvention | Beispiel
--- | --- | ---
Variablen | alles klein, wobei Leerzeichen durch `_` ersetzt werden | `my_variable`
konstante Variablen | alles gro geschrieben ' 
Funktionen | alles klein, wobei Leerzeichen durch `_` ersetzt werden | `my_function`
Klassen | CamelCase | `MyClass`
